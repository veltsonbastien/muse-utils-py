```python
import csv 

# Define the filename for the CSV
filename = 'agencies.csv'

# Open the CSV file in write mode
with open(filename, 'w', newline='') as file:
    # Create a CSV writer object
    writer = csv.writer(file)

    # Write the header row
    writer.writerow(['User Type', 'Phone', 'Twitter Handle', 'Company', 'Annual Revenue', 'Company Size', 'ICP Score', 'City', 'Country', 'State', 'Name', 'Email'])

    # Write the data rows
    for entry in agencies:
        groups, name, email = entry

        # Extract the individual fields from the groups list
        user_type = phone = twitterhandle = company = annualrevenue = company_size = icp_score = city = country = state = ''

        for group in groups:
            if group.startswith('user_type-'):
                user_type = group.replace('user_type-', '')
            elif group.startswith('phone-'):
                phone = group.replace('phone-', '')
            elif group.startswith('twitterhandle-'):
                twitterhandle = group.replace('twitterhandle-', '')
            elif group.startswith('company-'):
                company = group.replace('company-', '')
            elif group.startswith('annualrevenue-'):
                annualrevenue = group.replace('annualrevenue-', '')
            elif group.startswith('company_size-'):
                company_size = group.replace('company_size-', '')
            elif group.startswith('icp_score-'):
                icp_score = group.replace('icp_score-', '')
            elif group.startswith('city-'):
                city = group.replace('city-', '')
            elif group.startswith('country-'):
                country = group.replace('country-', '')
            elif group.startswith('state-'):
                state = group.replace('state-', '')

        # Write the row
        writer.writerow([user_type, phone, twitterhandle, company, annualrevenue, company_size, icp_score, city, country, state, name, email])
        
print(f"CSV file '{filename}' has been generated")
```
